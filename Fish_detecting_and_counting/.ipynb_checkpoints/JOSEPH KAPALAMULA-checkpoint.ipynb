# Data Cleaning in image dataset
1. **checking duplicate images and removing them**
duplicate images are images that coppy of other images in the same dataset,which can occur due to various reasons like repeated downloads, file migrations, or mistakes during data collection. removing duplicate images has he following advantages: increased data quality,improve training of a model and reduced storage usage.

 **here is a code to check for duplicate and remove them then creating new folder for filtered images**
 but before running the code we need to install library by running this **pip install Pillow imagehash** that is used to check for duplicate

```python
pip install Pillow imagehash
```

```python
import os
import shutil
from PIL import Image
import imagehash

def find_and_copy_unique_images(source_folder, filtered_folder):
    # Creating filtered folder
    if not os.path.exists(filtered_folder):
        os.makedirs(filtered_folder)

    image_hashes = {}

    for root, _, files in os.walk(source_folder):
        for file in files:
            if file.lower().endswith(('png', 'jpg', 'jpeg', 'gif', 'bmp')):
                image_path = os.path.join(root, file)
                try:
                    image = Image.open(image_path)
                    image_hash = imagehash.average_hash(image)
                    if image_hash in image_hashes:
                        print(f"Duplicate found: {file}, not copying it.")
                    else:
                        image_hashes[image_hash] = image_path
                        shutil.copy(image_path, os.path.join(filtered_folder, file))
                except Exception as e:
                    print(f"Error processing {file}: {e}")

source_folder1 = 'All_fish_Images'
filtered_folder = 'Filtered_images'
find_and_copy_unique_images(source_folder1, filtered_folder)
```

2.**resizing all images **
- resizing all images in the dataset to have 1024 by 768 pixels. Resizing ensures that all images in the dataset have same dimensions. This uniformity simplifies data preprocessing and model training, as the model does not have to handle images of different sizes.
- **resizing code of all images**

```python
import os
from PIL import Image

def resize_images(source_folder, width, height):
    total_images = 0
    
    # Create a folder to save resized images outside the source folder
    parent_folder = os.path.dirname(source_folder)
    resized_folder = os.path.join(parent_folder, "resized_images_new_folder2")#new folder name "resized_images_new_folder" will be created
    os.makedirs(resized_folder, exist_ok=True)

    for root, _, files in os.walk(source_folder):
        for file in files:
            if file.lower().endswith(('png', 'jpg', 'jpeg', 'gif', 'bmp')):
                image_path = os.path.join(root, file)
                try:
                    img = Image.open(image_path)
                    img = img.resize((width, height), Image.ANTIALIAS)
                    img.save(os.path.join(resized_folder, file))
                    total_images += 1
                except Exception as e:
                    print(f"Error processing {file}: {e}")

    return total_images


source_folder2 = 'Filtered_images'


width, height = 1024, 768
total_images = resize_images(source_folder2, width, height)
print(f"Total images processed: {total_images}")
```

3. **Removing Corrupted Images**
Removing corrupted images involves detecting and eliminating images that cannot be opened or processed. These images might be incomplete, have incorrect file headers, or be in a damaged state.

 **here is the code doing removing corrupted images** 

```python
import os
from PIL import Image

def remove_corrupted_images(source_folder):
    corrupted_folder = os.path.join(source_folder, "corrupted")
    if not os.path.exists(corrupted_folder):
        os.makedirs(corrupted_folder)
    
    corrupted_found = False

    for root, _, files in os.walk(source_folder):
        for file in files:
            if file.lower().endswith(('png', 'jpg', 'jpeg', 'gif', 'bmp')):
                image_path = os.path.join(root, file)
                try:
                    img = Image.open(image_path)
                    img.verify()  # Verify that this is an image file
                except (IOError, SyntaxError) as e:
                    print(f"Removing corrupted image: {file}")
                    os.rename(image_path, os.path.join(corrupted_folder, file))
                    corrupted_found = True
    
    if not corrupted_found:
        print("No corrupted images found.")

# Example usage
source_folder2 = 'resized'
remove_corrupted_images(source_folder2)
```

4. **Converting Image Formats into one format**
Converting image formats involves changing the file format of images within a dataset to ensure all images are in a consistent format. images where in this formats: 'png', 'jpg', 'jpeg', 'gif', 'bmp' then I have changed all format into **jpg**. 

**here is the code to convert image formats into one format in resize folder and code will create new folder containing images of same format**.

```python
import os
from PIL import Image

# Function to convert images to .jpg
def convert_images_to_jpg(source_folder, destination_folder):
    # Check if the source folder exists
    if not os.path.exists(source_folder):
        print(f"Source folder {source_folder} does not exist.")
        return
    
    # Create the destination folder if it does not exist
    os.makedirs(destination_folder, exist_ok=True)

    # Process each file in the source folder
    for filename in os.listdir(source_folder):
        # Get the full path of the file
        file_path = os.path.join(source_folder, filename)
        
        # Check if the file is an image
        try:
            with Image.open(file_path) as img:
                # Convert the image to RGB (required for JPEG format)
                rgb_img = img.convert('RGB')
                
                # Create a new filename with .jpg extension
                new_filename = os.path.splitext(filename)[0] + '.jpg'
                new_file_path = os.path.join(destination_folder, new_filename)
                
                # Save the image in JPEG format
                rgb_img.save(new_file_path, format='JPEG')
                print(f"Converted {filename} to {new_filename}")
        except Exception as e:
            print(f"Failed to process {filename}: {e}")

# Example usage
source_folder = source_folder2 
destination_folder = 'ALL_images in jpg FORMAT_new' # all converted images will be saved to this folder 'ALL_images in jpg FORMAT_new'
convert_images_to_jpg(source_folder, destination_folder)
```

5.**Removing Background Noise**,it enhence image quality by eliminating unnecessary elements in the background of the image.

**here is the code** but before running it you have to install opencv-python 

**NOTE**: when running the code to remove background color it makes the computer missbehaving i think it uses more computing power and other codes below it might work well when this code is runing


```python
import cv2
import numpy as np
import os

def remove_background(source_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for root, _, files in os.walk(source_folder):
        for file in files:
            if file.lower().endswith(('png', 'jpg', 'jpeg', 'gif', 'bmp')):
                image_path = os.path.join(root, file)
                img = cv2.imread(image_path)
                mask = np.zeros(img.shape[:2], np.uint8)

                bgdModel = np.zeros((1, 65), np.float64)
                fgdModel = np.zeros((1, 65), np.float64)

                rect = (50, 50, img.shape[1] - 50, img.shape[0] - 50)
                cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

                mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
                img = img * mask2[:, :, np.newaxis]

                output_path = os.path.join(output_folder, file)
                cv2.imwrite(output_path, img)

# Example usage
source_folder = 'ALL_images in jpg FORMAT'
output_folder = 'images with background removed'
remove_background(source_folder, output_folder)
```

# TRANSFORMING CLEANED IMAGES
1. **Rotating other images in the dataset**

Rotating an image means turning it around a center point, usually in increments of 90 degrees.  model generalisation.I have only rotated 100 pictures leaving other pictures so that model generalisation can be increased.
** 

i have rotated only hundred pictures inside the same folder named:'ALL_images in jpg FORMAT_new' 

```python
import pandas as pd
import os
from PIL import Image

def rotate_images_in_place(source_folder, angle=90, limit=100):
    count = 0

    for root, _, files in os.walk(source_folder):
        for file in files:
            if file.lower().endswith(('png', 'jpg', 'jpeg', 'gif', 'bmp')):
                if count >= limit:
                    return
                image_path = os.path.join(root, file)
                try:
                    img = Image.open(image_path)
                    img_rotated = img.rotate(angle, expand=True)
                    img_rotated.save(image_path)
                    count += 1
                    print(f"Rotated image: {file}")
                except Exception as e:
                    print(f"Error processing {file}: {e}")


source_folder = 'ALL_images in jpg FORMAT_new' 
rotate_images_in_place(source_folder, angle=90, limit=100)
```

# DATA FORMAT CHANGING FROM CSV TO COCO_JSON FORMAT
- converting the **csv** format I saved during annotating to **COCO_JSON** format using the python code 

- I changed the format to COCO_JSON format because it provides standardized and widely adopted way to present object detection,segmentation and keypoint annotation task.

# code used to convert CSV format to COCO_JSON format


```python
import json

def convert_to_coco(df):
    coco_format = {
        "images": [],
        "annotations": [],
        "categories": [
            {"id": 1, "name": "Fish"}
        ]
    }
    
    image_id = 1
    annotation_id = 1
    filenames = {}

    for _, row in df.iterrows():
        if row['filename'] not in filenames:
            filenames[row['filename']] = image_id
            coco_format['images'].append({
                "id": image_id,
                "file_name": row['filename'],
                "height": 0,  # Add actual height if known
                "width": 0    # Add actual width if known
            })
            image_id += 1

        shape_attributes = row['region_shape_attributes']
        bbox = [
            shape_attributes.get('x', 0),
            shape_attributes.get('y', 0),
            shape_attributes.get('width', 0),
            shape_attributes.get('height', 0)
        ]
        area = bbox[2] * bbox[3]  # width * height

        coco_format['annotations'].append({
            "id": annotation_id,
            "image_id": filenames[row['filename']],
            "category_id": 1,
            "bbox": bbox,
            "area": area,
            "iscrowd": 0
        })
        
        annotation_id += 1

    return coco_format

coco_data = convert_to_coco(df)

# Save to a COCO JSON file
with open('coco_annotations.json', 'w') as f:
    json.dump(coco_data, f)
print("done")
```
