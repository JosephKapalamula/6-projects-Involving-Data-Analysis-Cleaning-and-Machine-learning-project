
# Dealing with outliers
# visualising outliers using boxplot
# using "z-score method to find upper limit and lower limit to remove outliers"


```python
# Dealing with outliers in "MEAN"  and in "AREA" column becuase they are only column that that are numerical

# TO see outliers within the column we can visualise using boxplot
data4=data3.copy()
import seaborn as sns
import numpy as np

# TO see outliers within the column we can visualize using boxplot
sns.boxplot(data4["MEAN"]) # plot is showing that there's no anny outlier


# using "z-score method to find upper limit and lower limit to remove outliers", here we use mean and standard deviation

# finding upper limit of the column "MEAN"
upper_limit_in_Mean = data4["MEAN"].mean() + 3 * data4["MEAN"].std()
print(f"upper limit= {upper_limit_in_Mean}")

# finding lower limit of the column "MEAN"
lower_limit_in_Mean = data4["MEAN"].mean() - 3 * data4["MEAN"].std()
print(f"lower limit= {lower_limit_in_Mean}")

# Replace outliers with the mean of the column
data4["MEAN"] = np.where(data4["MEAN"] > upper_limit_in_Mean, data4["MEAN"].mean(), data4["MEAN"])
data4["MEAN"] = np.where(data4["MEAN"] < lower_limit_in_Mean, data4["MEAN"].mean(), data4["MEAN"])

```

```python
# Boxplot after removing outliers
sns.boxplot(data4["MEAN"]) # plots have no any outliers 
```

```python
# TO see outliers within the column of "AREA"
sns.boxplot(data4["AREA"]) # plot is showing that there's no anny outlier


# using "z-score method to find upper limit and lower limit to remove outliers", here we use mean and standard deviation

# finding upper limit of the column "AREA"
upper_limit_in_AREA = data4["AREA"].mean() + 3 * data4["AREA"].std()
print(f"upper limit_IN AREA= {upper_limit_in_AREA}")

# finding lower limit of the column "AREA"
lower_limit_in_AREA = data4["MEAN"].mean() - 3 * data4["MEAN"].std()
print(f"lower limit_IN AREA_COLUMN= {lower_limit_in_AREA}")

# Replace outliers with the mean of the column
data4["AREA"] = np.where(data4["AREA"] > upper_limit_in_Mean, data4["AREA"].mean(), data4["AREA"])
data4["AREA"] = np.where(data4["AREA"] < lower_limit_in_Mean, data4["AREA"].mean(), data4["AREA"])

```
